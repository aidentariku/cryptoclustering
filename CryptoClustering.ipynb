```python
# Import necessary libraries
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(r"C:\Users\AidenTariku\Downloads\Starter_Code (13)\Starter_Code\Resources\crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head()
```

```python
# Display summary statistics
df_market_data.describe()
```

```python
# Normalize the data using StandardScaler from scikit-learn
scaled_data = StandardScaler().fit_transform(df_market_data)
```

```python
# Create a DataFrame with the scaled data
df_scaled_market_data = pd.DataFrame(
    scaled_data,
    columns=df_market_data.columns
)

# Copy the crypto names from the original data
df_scaled_market_data["coin_id"] = df_market_data.index

# Set the coinid column as the index
df_scaled_market_data = df_scaled_market_data.set_index("coin_id")

# Display sample data
df_scaled_market_data.head()
```

```python
# Define a list containing the range of k-values to try
k_values = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia_values = []

# Iterate through each possible value of k to compute the inertia
for k_value in k_values:
    # Create a KMeans model using the current value of k
    kmeans_model = KMeans(n_clusters=k_value, random_state=0)
    
    # Fit the model to the scaled data
    kmeans_model.fit(df_scaled_market_data)
    
    # Append the inertia to the inertia_values list
    inertia_values.append(kmeans_model.inertia_)
```

```python
# Create a dictionary with the data for plotting the Elbow curve
elbow_curve_data = {
    "k_values": k_values,
    "inertia_values": inertia_values
}

# Create a DataFrame with the data for plotting the Elbow curve
df_elbow_curve = pd.DataFrame(elbow_curve_data)
```

```python
# Plot a line chart to visualize the inertia values for different values of k
elbow_curve_plot = df_elbow_curve.hvplot.line(x="k_values", y="inertia_values", title="Elbow Curve", xticks=k_values)
elbow_curve_plot
```

```python
# Initialize the K-Means model using the optimal value for k
kmeans_model = KMeans(n_clusters=4)
```

```python
# Fit the K-Means model to the scaled data
kmeans_model.fit(df_scaled_market_data)
```

```python
# Use the trained K-Means model to predict the clusters for the cryptocurrencies using the scaled data
crypto_clusters = kmeans_model.predict(df_scaled_market_data)

# Display the resulting array of cluster values
print(crypto_clusters)
```

```python
# Create a copy of the scaled DataFrame to store predictions
df_scaled_market_data_predictions = df_scaled_market_data.copy()
```

```python
# Add a new column to the DataFrame to store the predicted clusters
df_scaled_market_data_predictions["crypto_cluster"] = crypto_clusters

# Display sample data
df_scaled_market_data_predictions.head()
```

```python
# Create a scatter plot using hvPlot
clusters_plot = df_scaled_market_data_predictions.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="crypto_cluster",
    hover_cols=["coin_id"],
    marker=["hex", "square", "cross", "inverted_triangle"],
    title="Cryptocurrencies Clusters",
)
clusters_plot
```

```python
# Create a PCA model instance with n_components set to 3
pca_model = PCA(n_components=3)
```

```python
# Use the PCA model to transform the scaled market data to three principal components
market_pca_data = pca_model.fit_transform(df_scaled_market_data)

# View the first five rows of the transformed DataFrame
market_pca_data[:5]
```

```python
# Retrieve the explained variance ratio to determine the proportion of variance explained by each principal component
explained_variance_ratio = pca_model.explained_variance_ratio_
explained_variance_ratio
```

```python
# Creating a DataFrame with the PCA data
df_market_data_pca = pd.DataFrame(
    market_pca_data, 
    columns=["PC1", "PC2", "PC3"]
)

# Copy the crypto names from the original data
df_market_data_pca["coin_id"] = df_market_data.index

# Set the coinid column as index
df_market_data_pca = df_market_data_pca.set_index("coin_id")

# Display sample data
df_market_data_pca.head()
```

```python
# Create a list with the number of k-values to try
# Use a range from 1 to 10
k = list(range(1, 11))
```

```python
# Create an empty list to hold the inertia values
inertia_values = []

# Iterate through each potential value of k to calculate the inertia
# Within the loop:
# 1. Initialize a KMeans model with the current value of k
# 2. Fit the model to the PCA-transformed data using `df_market_data_pca`
# 3. Append the inertia value from the model to the inertia_values list
for i in k:
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(df_market_data_pca)
    inertia_values.append(kmeans.inertia_)
```

```python
# Prepare a dictionary containing the data for generating the Elbow curve
elbow_curve_data_pca = {
    "k_values": k,
    "inertia_values": inertia
}

# Create a DataFrame to facilitate plotting the Elbow curve
df_elbow_curve_pca = pd.DataFrame(elbow_curve_data_pca)
```

```python
# Generate a line chart displaying the inertia values for different values of k to identify the optimal k
elbow_curve_plot_pca = df_elbow_curve_pca.hvplot.line(x="k_values", y="inertia_values", title="Elbow Curve Using PCA Data", xticks=k)
elbow_curve_plot_pca
```

```python
## Initialize the K-Means model with the chosen optimal value for k
kmeans_model_pca = KMeans(n_clusters=4)
```

```python
# Fit the K-Means model to the PCA-transformed data
kmeans_model_pca.fit(df_market_data_pca)
```

```python
# Use the trained K-Means model to predict the clusters for the cryptocurrencies using the PCA-transformed data
crypto_clusters_pca = kmeans_model_pca.predict(df_market_data_pca)

# Display the resulting array of cluster values
print(crypto_clusters_pca)
```

```python
# Create a copy of the DataFrame containing the PCA data
df_market_data_pca_predictions = df_market_data_pca.copy()

# Add a new column to the DataFrame to store the predicted clusters
df_market_data_pca_predictions["crypto_cluster"] = crypto_clusters_pca

# Display sample data
df_market_data_pca_predictions.head()
```

```python
# Create a scatter plot using hvPlot with PC1 and PC2 as x and y axes respectively
# Color the points based on the predicted clusters obtained from K-Means
# Include the cryptocurrency name in the hover information
clusters_plot_pca = df_market_data_pca_predictions.hvplot.scatter(
    x="PC1",
    y="PC2",
    by="crypto_cluster",
    hover_cols=["coin_id"],
    marker=["hex", "square", "cross", "inverted_triangle"],
    title="Cryptocurrency Clusters Using PCA Data"
)
clusters_plot_pca
```

```python
# Composite plot to compare the Elbow curves for original data and PCA-transformed data
composite_elbow_plot = elbow_plot * elbow_plot_pca
composite_elbow_plot
```

```python
# Composite plot to compare the clusters for original data and PCA-transformed data
composite_clusters_plot = clusters_plot * clusters_plot_pca
composite_clusters_plot
```
